```python
import pandas as pd
df1 = pd.read_csv("counties1.csv")
df2 = pd.read_csv("counties2.csv")
df3 = pd.read_csv("counties3.csv")
df4 = pd.read_csv("counties4.csv")
df5 = pd.read_csv("counties5.csv")
dic = pd.read_csv("dictionary.csv")
```

```python
varnames = dic['varname']

df1 = df1[0:62]
df1 = df1[df1.columns.drop(list(df1.filter(regex='Fact')))]
df1 = df1[df1.columns.drop(list(df1.filter(regex='Value')))]
dfnew1 = df1.transpose()
dfnew1.columns = varnames

df2 = df2[0:62]
df2 = df2[df2.columns.drop(list(df2.filter(regex='Fact')))]
df2 = df2[df2.columns.drop(list(df2.filter(regex='Value')))]
dfnew2 = df2.transpose()
dfnew2.columns = varnames

df3 = df3[0:62]
df3 = df3[df3.columns.drop(list(df3.filter(regex='Fact')))]
df3 = df3[df3.columns.drop(list(df3.filter(regex='Value')))]
dfnew3 = df3.transpose()
dfnew3.columns = varnames

df4 = df4[0:62]
df4 = df4[df4.columns.drop(list(df4.filter(regex='Fact')))]
df4 = df4[df4.columns.drop(list(df4.filter(regex='Value')))]
dfnew4 = df4.transpose()
dfnew4.columns = varnames

df5 = df5[0:62]
df5 = df5[df5.columns.drop(list(df5.filter(regex='Fact')))]
df5 = df5[df5.columns.drop(list(df5.filter(regex='Value')))]
dfnew5 = df5.transpose()
dfnew5.columns = varnames
```

```python
frames = [dfnew1, dfnew2, dfnew3,dfnew4,dfnew5]
df = pd.concat(frames)
```
